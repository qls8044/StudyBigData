```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup_items = soup.select('items > item')
```

```python
len(soup_items)
```

```python
import pandas as pd
```

```python
item_book = pd.DataFrame()
```

```python
for idx, item in enumerate(soup_items):
    cinaddr1 = ''
    cinaddr2 = ''
    ciname = ''
    
    try:
        if len(item.select('cinaddr1')) > 0:
            cinaddr1 = item.select('cinaddr1')[0].text
        if len(item.select('cinaddr2')) > 0:
            cinaddr2 = item.select('cinaddr2')[0].text
            
        ciname = item.select('ciname')[0].text
    except:
        continue

    book = pd.DataFrame({ 'cinaddr1': [cinaddr1]
                         ,'cinaddr2': [cinaddr2]
                         , 'ciname': [ciname] })

    item_book = item_book.append(book)
```
