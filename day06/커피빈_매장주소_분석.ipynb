```python
# 라이브러리 추가할 때 이 셀에 추가할 것임
import pandas as pd 
import folium
```

```python
cb_data = pd.read_csv('./data/CoffeeBean.csv', index_col=0 ,header=0, encoding='utf-8')
cb_data.head(20)
```

```python
# 현재 주소 체계 확인
addr = []

for address in cb_data.address:
    addr.append(str(address).split())
addr
```

```python
addr2 = []

for i in range(len(addr)):
    if addr[i][0] == '서울': addr[i][0] = '서울특별시'
    elif addr[i][0] == '서울시': addr[i][0] = '서울특별시'
    elif addr[i][0] == '부산시': addr[i][0] = '부산광역시'
    elif addr[i][0] == '인천': addr[i][0] = '인천광역시'
    elif addr[i][0] == '광주': addr[i][0] = '광주광역시'
    elif addr[i][0] == '대전시': addr[i][0] = '대전광역시'
    elif addr[i][0] == '울산시': addr[i][0] = '울산광역시'
    elif addr[i][0] == '세종시': addr[i][0] = '세종특별자치시'
    elif addr[i][0] == '경기': addr[i][0] = '경기도'
    elif addr[i][0] == '충북': addr[i][0] = '충청북도'
    elif addr[i][0] == '충남': addr[i][0] = '충청남도'
    elif addr[i][0] == '전북': addr[i][0] = '전라북도'
    elif addr[i][0] == '전남': addr[i][0] = '전라남도'
    elif addr[i][0] == '경북': addr[i][0] = '경상북도'
    elif addr[i][0] == '경남': addr[i][0] = '경상남도'
    elif addr[i][0] == '제주도': addr[i][0] = '제주특별자치도'
    elif addr[i][0] == '제주시': addr[i][0] = '제주특별자치도'
    elif addr[i][0] == '제주': addr[i][0] = '제주특별자치도'

    addr2.append(' '.join(addr[i])) # 공백을 기준으로 다시 합침

addr2
```

```python
# addr2를 DataFrame으로 변경
addr2 = pd.DataFrame(addr2, columns=['address2'])
```

```python
cb_data2 = pd.concat([cb_data, addr2], axis=1) # axis=1은 컬럼으로 concat하는 것임(axis='columns' 로도 쓸 수 있음)
cb_data2.head()
```

```python
# 중간저장
cb_data2.to_csv('./data/CoffeeBean_1.csv', encoding='utf-8', index=False)
```

### folium 사용법

```python
map = folium.Map(location=[35.172801,129.130628], zoom_start=15, tiles='Stamen Terrain')
map
```

```python
map = folium.Map(location=[35.172801,129.130628], zoom_start=15)
folium.Marker(location=[35.172801,129.130628], popup='부산정보산업진흥원',icon=folium.Icon(color='purple', icon='home')).add_to(map)
# icon 그림은 icon=''에 찾아서 넣으면 됨. 구글에 folium marker icon해서 찾아보면 됨. home, remove, ok 등등 있음
# https://getbootstrap.com/docs/3.3/components/ 여기서 marker icon 원하는거 골라쓰면 됨
map
```

```python
# html 저장
map.save('./data/map.html')
```
